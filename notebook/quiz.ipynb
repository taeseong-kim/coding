#### 2) 체감온도 및 불쾌지수 추가
- 불쾌지수
    - 0.81 * 섭씨온도 +0.01 * 상대습도(%)（0.99*섭씨온도 - 14.3）+ 46.3



```python
### 코드는 여기에!!### 
## 함수 만들어서 해주세요 ~~
# jeju_data 를 활용해주세요! 
# 그냥하면 1점 감점 
```

```python
# code 는 여기에! 
```

```python
ls = []
def degree(info):
    
    for temp,humid in info:
    
        result = 0.81*temp + 0.01*humid*(0.99*temp - 14.3) + 46.3
        ls.append(result)
```

```python
a = jeju_data["temp"].values
b = jeju_data["humid"].values
```

```python
degree(list(zip(a, b)))
```

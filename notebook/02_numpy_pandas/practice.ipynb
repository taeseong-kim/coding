```python
df_1 = df.pivot_table("count", "RANK", "SEX", aggfunc="sum")
df_1.head()
```

```python
count = df.groupby(["SEX"]).size()
```

```python
result = df_1/count * 100
```

```python
df["POINT"] = df["RANK"].apply(
    lambda rank: 4-rank if rank <= 3 else 0
)
```

```python
df_2 = df[["RCHOSE_NM", "POINT"]]
```

```python
result = df_2.groupby("RCHOSE_NM").agg("sum")
result = result.sort_values("POINT", ascending=False)
result.reset_index(inplace=True)
```
